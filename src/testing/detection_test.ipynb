```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow  # Import Colab-specific imshow

# Load your trained YOLOv11 model
model = YOLO("yolov11s_50e_32s_visdrone.pt")  # Replace with the path to your model.pt file

# Load the image
image_path = "car_test.jpg"  # Replace with your image file path
img = cv2.imread(image_path)

# Run inference
results = model(img)

# Display the results
annotated_img = results[0].plot()  # This adds bounding boxes and labels to the image
cv2_imshow(annotated_img)  # Use Colab's cv2_imshow instead of cv2.imshow

# Optionally, save the result
cv2.imwrite("output_image.jpg", annotated_img)  # Save the annotated image
```
